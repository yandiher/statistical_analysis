Diketahui:
Seorang tukang cilung, dalam waktu 100 hari, mampu menjual 200 sampai 250 cilung per hari.
Ditanyakan:
Berapa tusuk cilung yang perlu disiapkan oleh tukang cilung per hari agar, jika tidak habis dagangannya, tidak terlalu rugi.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
n = np.random.randint(low=200, high=250, size=100)
```

n adalah ilustrasi data yang nantinya akan diganti oleh data yang sebenarnya

```python
mean = np.mean(n)
median = np.median(n)
std = np.std(n)
```

```python
plt.figure(figsize=(16,5))
sns.boxplot(x=n)
plt.axvline(x=mean, color='r', label='mean')
plt.title('boxplot untuk visualisasi data')
plt.legend()
plt.show()
```
